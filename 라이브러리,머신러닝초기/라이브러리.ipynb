```python
def remove_punc_1(li_string):
    li_temp = []
    for string in li_string:
        string = re.sub('[!~]',"",string)
        string = string.strip()
        li_temp.append(string)
    return li_temp
remove_punc_1(state) # 리스트로는 안되니까 포문 써준다. 
```

```python
# 2 모든 연도 파일  합치기 
years = range(1880,2011) # 년도의 범위
pieces = [] # 빈칸 어펜드 시킬 것 
columns = ['name','sex','birth'] # 열 네임 
# 이어부터 이어스 까지 1880~ 2011
# 페스를 반복할 꺼야 근데 숫자가 year에 따라서 바뀔꺼야
# 프레임은 페스를 읽는 거야 
# 프레임에서 year 인덱스는 year에서 오는 거야 
# 그리고 그 값을 피스에 넣어줄거야
#그리고 그 피스들을 합쳐줄거야 
for year in years: 
    path = 'C:/python/pydata-book-2nd-edition/pydata-book-2nd-edition/datasets/babynames/yob%d.txt'% year
    frame = pd.read_csv(path,names = columns)
    frame['year'] = year
    pieces.append(frame)
    
names = pd.concat(pieces,ignore_index=True)
```

```python
# 3그룹바이나 피벗테이블로 묶기 그리고 시각화 
# 피벗테이블로 만들기 출생자수는 기준 행은 년도 열은 성별 합으로 구한다. 
total_births = names.pivot_table('birth',index = 'year',columns = 'sex', aggfunc=sum)

total_births.tail()
total_births.plot(title = 'total birth by sex and year')
```

```python
# 4 인기 있는 이름 
# 다 콘켓시킨 것이 네임인데 거기에서 년도와 성별 기준으로 묶은다음 전체 출생에서 그 그룹의 출생의 비율을 알아보는 함수를 대입
def add_prop(group):
    group['prop']= group.birth/group.birth.sum()
    return group
names = names.groupby(['year','sex']).apply(add_prop)
```
