```python
iris = np.loadtxt('iris.csv', skiprows=1,
                 delimiter=',', usecols=range(4))
iris.shape
```

```python
sl = iris[:, 0]
sl
```

```python
m = sl.mean()
s = sl.std()
m, s
```

```python
(sl>(m+s)).sum() + (sl<(m-s)).sum()
```

```python
(np.abs(sl-m)>s).sum()
```
