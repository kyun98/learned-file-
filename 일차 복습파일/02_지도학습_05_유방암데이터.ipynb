### 상관계수
- Pearson correlation coefficient
- https://en.wikipedia.org/wiki/Pearson_correlation_coefficient

$$ corr(X,Y) = \frac{\sum_{i=0}^{N-1} (X_i - \bar{X}) \cdot (Y_i - \bar{Y})} {N \sigma_x \sigma_y} $$
(sigma 는 표준편차를 의미함)

#### 연습문제)
- 유방암 데이터에서 0번 속성과 3번 속성의 상관계수를 위의 공식으로 구하시오.
- 위 데이터에 선형회귀를 적용하여 기울기와 절편값을 구하시오. 그리고 산점도와 구한 직선을 그리시오.

```python
X = cancer.data[:,0]
Y = cancer.data[:,3]
X.shape, Y.shape
```

```python
m_x = X.mean()
m_y = Y.mean()
s_x = X.std()
s_y = Y.std()
```

```python
((X-m_x)*(Y-m_y)).mean()/(s_x*s_y)
```

### 정규화 적용 (k-NN, LinearSVC, LogisticRegression)
- train_test_split() 적용 후 score 확인
- 원본 데이터와 정규화된 데이터 비교
- 10번 반복해서 Logistic Regression 과 선형 SVM 을 적용해 보자. train_test_split() 함수에서 랜덤하게 데이터를 나누기 때문에 매번 점수가 달라진다.

- kNN

```python
from sklearn.neighbors import KNeighborsClassifier

scores = []

for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target)

    model = KNeighborsClassifier()
    model.fit(X_train,y_train)

    score = model.score(X_test,y_test)
    scores.append(score)

print('scores =', scores, '\nmean =', sum(scores)/10)
```

```python
m = cancer.data.mean(axis=0)
s = cancer.data.std(axis=0)
data2 = (cancer.data - m) / s

scores = []

for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(data2, cancer.target)

    model = KNeighborsClassifier()
    model.fit(X_train,y_train)

    score = model.score(X_test,y_test)
    scores.append(score)

print('scores =', scores, ', mean =', sum(scores)/10)
```

- LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

scores = []

for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target)

    model = LogisticRegression(C=1)
    model.fit(X_train,y_train)

    score = model.score(X_test,y_test)
    scores.append(score)

print('scores =', scores, ', mean =', sum(scores)/10)
```

```python
scores = []

for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(data2,cancer.target)

    model = LogisticRegression(C=1)
    model.fit(X_train,y_train)

    score = model.score(X_test,y_test)
    scores.append(score)

print('scores =', scores, ', mean =', sum(scores)/10)
```

- LinearSVC

```python
from sklearn.svm import LinearSVC

scores = []

for i in range(10):
    X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target)

    model = LinearSVC()
    model.fit(X_train,y_train)

    score = model.score(X_test,y_test)
    scores.append(score)

print('scores =', scores, ', mean =', sum(scores)/10)
```
