- sigmoid 
$$ y = {{1} \over {1 + e^{-x}} }$$

```python
x = np.arange(-10, 10, 0.1)
y = 1/(1+np.exp(-x))
plt.plot(x, y)
#plt.axis('equal')
```

```python
x = np.arange(-10, 10, 0.1)
y = np.where(x<0, 0, x)
plt.plot(x, y)
#plt.axis('equal')
```

- relu 

```python
x = np.arange(-10, 10, 0.1)
y = x.copy(); y[y<0] = 0
plt.plot(x, y)
#plt.axis('equal')
```

```python
x = np.arange(-1, 1, 0.01)
y = np.sqrt(-x**2+1)
plt.plot(x, y)
plt.axis('equal')
```
