```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
X = cancer.data
X.shape
```

```python
y = cancer.target
```

```python
X_norm = ((X - X.mean(axis=0)) / X.std(axis=0))[:,:2]
X_norm.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_norm, y)
```

```python
model.score(X_norm,y)
```

```python
import mglearn

mglearn.plots.plot_2d_separator(model, X_norm, fill=False, eps=0.5, alpha=0.4)
mglearn.discrete_scatter(X_norm[:, 0], X_norm[:, 1], y, alpha=0.5)
plt.legend(['malignant','benign'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
X = cancer.data[:, :3]
X.shape
```

```python
W1 = np.random.randn(3,3)
b1 = np.random.randn(3)

y1 = sigmoid(X@W1 + b1)
y1.shape
```

```python
W2 = np.random.randn(3,4)
b2 = np.random.randn(4)

y2 = sigmoid(y1@W2 + b2)
y2.shape
```

```python
W3 = np.random.randn(4,2)
b3 = np.random.randn(2)

pred_y = y2@W3 + b3
pred_y.shape
```
