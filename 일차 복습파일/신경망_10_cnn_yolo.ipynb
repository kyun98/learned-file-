### Keras 용 YOLO 설치
- https://github.com/qqwweee/keras-yolo3

#### 설치 절차
- 가상환경 생성 : conda create -n keras-yolo3 python=3.5
- 가상환경 실행 : activate keras-yolo3
- pip install tensorflow==1.6.0
- pip install keras==2.1.5
- pip install h5py, Pillow, matplotlib, opencv-python
<hr/>

#### 실행 절차
- https://github.com/qqwweee/keras-yolo3 깃허브 다운로드
- https://pjreddie.com/media/files/yolov3.weights 다운로드
- python convert.py yolov3.cfg yolov3.weights model_data/yolo.h5
- 이미지 검출 : python yolo_video.py --image
- 비디오 검출 : python yolo_video.py --input test.mp4 --output test_result.mp4
