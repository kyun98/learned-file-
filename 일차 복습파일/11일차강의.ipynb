### 연습문제
- 아이리스의 각 속성을 정규화하시오. (평균 0, 표준편차 1)
- 정규화한 데이터를 plt.plot() 으로 그리시오.
- 원본 데이터와 위에서 그린 그림을 하나의 그림에 나타내시오. (plt.subplot() 사용)

```python
iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',', usecols=range(4))
iris
```

```python
iris_norm = (iris - iris.mean(axis=0)) / iris.std(axis=0)
iris_norm
```

```python
plt.plot(iris_norm)
plt.legend(['sl','sw','pl','w'])
```

```python
plt.subplot(1,2,1)
plt.plot(iris)
plt.legend(['sl','sw','pl','w'])

plt.subplot(1,2,2)
plt.plot(iris_norm, alpha=0.5)
plt.legend(['sl','sw','pl','w'])
```

```python
plt.subplot(2,2,2)
plt.plot([1,2,3])

plt.subplot(2,2,3)
plt.plot([1,2,4,2])
```
